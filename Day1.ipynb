```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '/content/drive/MyDrive/Dataset/placement.csv'
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package (in lpa)')
plt.show()
```

At first I have to know which one is input and which one is output and for now x is input and y is output

```python
X = df.iloc[:, 0:1]
Y = df.iloc[:, -1]
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
X_test
```

```python
Y_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'] , df['package'])
plt.plot(X_train, lr.predict(X_train) , color='red')
plt.xlabel('CGPA')
plt.ylabel('Package (in lpa)')
plt.show()
```
